```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/company_sales_data.csv')
df
```

1st question

```python
x=df['month_number']
y=df['total_profit']
plt.plot(x,y)
plt.xlabel("month_ number")
plt.ylabel("total profit")
plt.xticks(x)
plt.yticks([200000,250000,300000,350000,400000,450000,500000,550000,600000])
plt.legend(loc='lower right')
```

2 nd question

```python
x=df['month_number']
y=df['total_profit']
plt.plot(x,y,linestyle='dotted',color='r',marker='o',linewidth=3,label='profit-data')
plt.xlabel("month_ number")
plt.ylabel("Sold units number")
plt.xticks(x)
plt.yticks([200000,250000,300000,350000,400000,450000,500000,550000,600000])
plt.legend(loc='lower right')
```

3 rd question

```python
facecream=df['facecream']
facewash=df['facewash']
toothpaste=df['toothpaste']
bathingsoap=df['bathingsoap']
moisturizer=df['moisturizer']
shampoo=df['shampoo']
plt.plot(x,facecream,marker='s',color='g',label='face cream sales')
plt.plot(x,facewash,marker='s',color='y',label='facewash sales')
plt.plot(x,toothpaste,marker='s',color='c',label='tooth_ paste sales')
plt.plot(x,bathingsoap,marker='s',color='b',label='bathig soap sales')
plt.plot(x,moisturizer,marker='s',color='k',label='moisturizer sales')
plt.plot(x,shampoo,marker='s',color='r',label='shampoo sales')
plt.xlabel('Month-number')
plt.ylabel('Total sales of each category')
plt.legend(loc='upper right')
plt.xticks(x)
plt.yticks([1000,2000,3000,4000,5000,6000,7000])



```

4th question

```python
plt.scatter(x,toothpaste)
```

5th question

```python
plt.bar([a-0.25 for a in x],facecream,width=0.25,align='edge',label='face_cream_sales',color='y')
plt.bar([a+0.25 for a in x],facewash,width=-0.25,align='edge',label='facewash_sales',color='r')
plt.tight_layout
plt.grid()
plt.xticks(x)
plt.legend(loc='upper right')
plt.xlabel('Month in number')
plt.ylabel('profit')
plt.title("face_cream-face_wash profit analysis")
```

6 th question

```python
total_profit=df['total_profit']
plt.hist(total_profit,bins=10,edgecolor='black')
```

7th question

```python
label=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
sales=[df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
plt.pie(sales,labels=label,autopct='%1.0f%%')
plt.legend(loc='lower right')
plt.tight_layout()
```
